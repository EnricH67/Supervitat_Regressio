```python
# defineixo el sample size que vull
N = 2000

# genero una nou dataset amb el random sample estratificat
sample_retards_df = retards_df .groupby('UniqueCarrier', group_keys=False).apply(lambda x: x.sample(int(np.rint(N*len(x)/len(retards_df))))).sample(frac=1).reset_index(drop=True)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
numeric_features = ['Month', 'DayofMonth', 'DayOfWeek', 'ActualElapsedTime', 'CRSElapsedTime', 'AirTime', 'Distance', 'TaxiIn',
                    'TaxiOut', 'CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay','LateAircraftDelay']
```

```python
ss = StandardScaler()
```

```python
sample_retards_df[['Month', 'DayofMonth', 'DayOfWeek', 'ActualElapsedTime', 'CRSElapsedTime', 'AirTime', 'Distance', 'TaxiIn',
                   'TaxiOut', 'CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay','LateAircraftDelay']] = ss.fit_transform(sample_retards_df[['Month', 'DayofMonth', 'DayOfWeek', 'ActualElapsedTime', 'CRSElapsedTime', 'AirTime',
 'Distance', 'TaxiIn', 'TaxiOut', 'CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay',
       'LateAircraftDelay']])
```

```python
sample_retards_df
```

```python
retards_dummies_df = pd.get_dummies(sample_retards_df, columns = ['UniqueCarrier','Cancelled', 'Diverted'])
retards_dummies_df
```

+ *Poso la variable target al final del DataFrame*

```python
a = retards_dummies_df.drop("ArrDelay", 1)
b = retards_dummies_df["ArrDelay"]
```

```python
retards_dummies_df = pd.concat([a, b], axis = 1)
```

```python
retards_dummies_df
```

```python
X = retards_dummies_df.drop("ArrDelay", 1)
y = retards_dummies_df["ArrDelay"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state = 4)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

###  MULTILINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

+ *L'entrenem amb ".fit" sobre les variables independents*

```python
model.fit(X_train, y_train)
```

+ *Obtenim R2 - coeficient de determinaciÃ³ amb X_train i y_train*

```python
r_sq = model.score(X_train, y_train)
print('coefficient of determination:', r_sq)
```

+ *Fem una predicciÃ³ amb ".predict(X_test)" - on X_test = variables independents NO conegudes pel model -*

```python
y_predict = model.predict(X_test)
y_predict
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print('MSE = {}'.format (mean_squared_error(y_test, y_predict)))
```

+ *El resultat ha estat  MOOOLT pitjor quan teniem la variable "DepDelay" fortament correlacionada amb la variable objectiu*
